## Loading the data
We can now load the dataset into pandas using the read_csv() function. This converts the CSV file into a Pandas dataframe.

```python
df = pd.read_csv("vip.csv")
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
en_ar = {
'أ' : 'A',
'ا' : 'A',
'ب' : 'B',
'ح' : 'J',
'د' : 'D',
'ر' : 'R',
'س' : 'S',
'ص' : 'X',
'ط' : 'T',
'ع' : 'E',
'ق' : 'G',
'ك' : 'K',
'ل': 'L',
'م' : 'Z',
'ن' : 'N',
'هـ' : 'H',
'ه' : 'H',
'و' : 'U',
'ى' : 'V'
}
```

```python
df
```

```python
df.rename(columns={'Arabic Letters':'char_ar','English Numbers':'plate_no'},inplace=True)
```

```python
a = df['char_ar'][0][0]
b = 'ل'
a == b
```

```python
import unicodedata

def normalize_arabic(text):
    return ' '.join(unicodedata.normalize('NFKC', char) for char in text)

df['char_ar'] = df['char_ar'].apply(normalize_arabic)
```

```python
df['char_ar'][1].split()[0]
```

```python
df['char1_ar'] = df['char_ar'].apply(lambda x: x.split()[0])
```

```python
df['char2_ar'] = df['char_ar'].apply(lambda x: x.split()[1])
```

```python
df['char3_ar'] = df['char_ar'].apply(lambda x: x.split()[2])
```

```python
df['char1_en'] = df['char1_ar'].apply(lambda x: en_ar[x])
```

```python
df['char2_en'] = df['char2_ar'].apply(lambda x: en_ar[x])
```

```python
df['char3_en'] = df['char3_ar'].apply(lambda x: en_ar[x])
```

```python
df.rename(columns={'Price':'price'},inplace=True)
```

```python
df['char_en'] = df['char3_en'] + ' ' + df['char2_en'] + ' ' + df['char1_en']
```

```python
df = df[['char_ar','char_en','char1_ar','char2_ar','char3_ar','char1_en','char2_en','char3_en','plate_no','price']]
```
