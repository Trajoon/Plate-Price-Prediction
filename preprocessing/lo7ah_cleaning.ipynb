```python
df = pd.concat([df, df2], ignore_index=True)
```

```python
df=df[df['chakra-text 16'] != 0 ]
```

```python
df = df.drop(columns=['Href','image','City','Publication date'] )
```

```python
df['char_ar'] = df['char1_ar']+ ' ' + df['char2_ar'] + ' ' + df['char3_ar']
```

```python
df['plate_no'] = df['Plate number']
```

```python
df['char_en'] = df['char3_en']+ ' ' +df['char2_en']+ ' ' +df['char1_en']
```
